# IPL Match Data Analysis

## Objective

The goal of this project is to analyze IPL match data and identify patterns that influence match outcomes.

## Dataset

The dataset used in this analysis is `ipl_matches_100.csv`, which contains historical IPL match information such as teams, venue, season, toss winner, and match winner.

```python
import pandas as pd
```

```python
df = pd.read_csv("ipl_matches_100.csv")
```

```python
df.head()
```

```python
df["match_winner"]
```

## 1. Most Successful Teams

```python
team_wins=df["match_winner"].value_counts()
team_wins
```

```python
team_wins.head()
```

### Analysis
The analysis shows that two teams CSK and DC have the highest number of wins (20 each) in the dataset. This indicates that these teams have been the most successful among the sampled IPL matches.

## 2. Toss Impact on Match Results

```python
df["toss_winner"] == df["match_winner"]
```

```python
toss_result=(df["toss_winner"] == df["match_winner"]).value_counts()
toss_result
```

```python
import matplotlib.pyplot as plt
```

```python
toss_result.plot(kind="pie", autopct="%1.1f%%")

plt.title("Impact of Toss on Match Results")

plt.show()
```

### Analysis
The results show that only 17% of matches were won by the team that also won the toss, while 83% were won by the opposing team. This suggests that winning the toss alone does not strongly determine the match outcome.

## 3. Venue Analysis

```python
venue_matches = df["venue"].value_counts()
venue_matches
```

```python
import matplotlib.pyplot as plt

venue_matches.plot(kind="bar", figsize=(10,5))
plt.title("Number of Matches Played at Each Venue")
plt.xlabel("Venue")
plt.ylabel("Number of Matches")
plt.show()
```

### Analysis
The venue analysis shows that some stadiums host more matches than others in the dataset. Venue conditions can influence match strategies because pitch behavior varies across different stadiums.

## 4. Season Trend Analysis

```python
season_matches = df["season"].value_counts().sort_index()
season_matches
```

```python
import matplotlib.pyplot as plt

season_matches.plot(kind="line", marker="o", figsize=(8,5))

plt.title("Number of Matches per IPL Season")
plt.xlabel("Season")
plt.ylabel("Number of Matches")

plt.show()
```

### Analysis
The season trend chart shows how the number of matches in the dataset varies across different IPL seasons. Some seasons contain more matches in the dataset than others, indicating uneven distribution of sampled matches.

## Strategic Insight

The analysis shows that toss advantage does not strongly determine match outcomes, as only 17% of toss-winning teams went on to win the match. This suggests that overall team performance and strategy play a larger role than toss advantage.
